```python
import sympy
from sympy.plotting import plot
```

```python
x = sympy.Symbol('x')
print(x)
```

```python
func = -12 * x**4 * sympy.sin(sympy.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
print(func)
```

```python
print(func.subs(x, 10.0))
```

```python
plot(func, ylim=(-100, 100), xlim=(-10, 10))
```

```python
#первый корень

for i in range (-765062232, -765062225):
    print (i/100000000, func.subs(x,i/100000000)) 
```
